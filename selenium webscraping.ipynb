### This webscraping file is to use selenium to scrape the diamonds data from www.bluenile.com. The webpages on this website has lot of button tags that you need to click to show more items (features), as well as input tags to screen/filter the goods. Also, the webpages need to scroll down to load more diamond items. So Selenium is better to do this job.

```python
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
def wait_tag(br, t, by, name):
    # to integrate the webdriverwait function 
    tag = False
    try:
        tag = WebDriverWait(br, t).until(EC.presence_of_element_located((by, name)))
        return tag
    except Exception as e:
        return tag
```

```python
def fs(s):
    # the function to find the contents of information
    if not s:
        return ''
    return str(s).replace('\t', '').replace('\n', '').replace('\xa0', ' ').strip().replace("'", '"')
```

```python
def get_br():
    # function to open the firefox browser 
    br = webdriver.Firefox(executable_path='./geckodriver', )
    br.get('https://www.bluenile.com/diamond-search')
   # br.get('https://www.bluenile.com/diamond-search')
    return br
```

```python
def enbale_sel(did):
    # to turn on the 6 buttons on webpage to show more features and for inputing values
    try:
        depth = br.find_element_by_css_selector('div[id="{}"]'.format(did, )).find_element_by_css_selector(
            'div[class="toggle-button-background"]')
        depth.click()
        sleep(2)
    except Exception as e:
        print('opened')
```

```python
def input_parm(name, kw):
    # the function to input the feature values of diamonds
    try:
        br.find_element_by_css_selector('input[name="{}"]'.format(name)).clear()
        sleep(1)
        br.find_element_by_css_selector('input[name="{}"]'.format(name)).send_keys(kw)
        sleep(1)
    except Exception as e:
        print('parameters not imput')
```

```python
# give the parameters values
parms = {
    'price-min-input': 730,
    'price-max-input': 1000000,
    'carat-min-input': 0.23,
    'carat-max-input': 20.98,
    'depth%-min-input': 51,
    'depth%-max-input': 74,
    'table%-min-input': 43,
    'table%-max-input': 79,
    'lengthwidthratio-min-input': 1.00,
    'lengthwidthratio-max-input': 1.62}
```

```python
br = get_br()
try:
    wait_tag(br, 20, By.CSS_SELECTOR, 'button[class="filters-overflow-button more"]').click()
except Exception as e:
    print('turn on more')

# to get more fields
buttons = br.find_elements_by_css_selector('div[class="filter-content add-column"]')
for but in buttons:
    but.click()
    sleep(2)

# turn on Polish
enbale_sel('polish-filter-container')

# turn on Symmetryeyi1
enbale_sel('symmetry-filter-container')

# turn on Fluorescence
enbale_sel('fluorescence-filter-container')

# turn on depth
enbale_sel('depth-percentage-filter-container')

# turn on table
enbale_sel('table-percentage-filter-container')

# turn on L/W
enbale_sel('length-to-width-ratio-filter-container')

# Price min input
input_parm('price-min-input', parms.get('price-min-input', 0))

# Price Max input
input_parm('price-max-input', parms.get('price-max-input', 0))

# carat min input
input_parm('carat-min-input', parms.get('carat-min-input', 0))

# carat max input
input_parm('carat-max-input', parms.get('carat-max-input', 0))

# depth min input
input_parm('depth%-min-input', parms.get('depth%-min-input', 0))

# depth max input
input_parm('depth%-max-input', parms.get('depth%-max-input', 0))

# table min input
input_parm('table%-min-input', parms.get('table%-min-input', 0))

# table max input
input_parm('table%-max-input', parms.get('table%-max-input', 0))

# lengthwidthratio  min input
input_parm('lengthwidthratio-min-input', parms.get('lengthwidthratio-min-input', 0))

# lengthwidthratio  max input
input_parm('lengthwidthratio-max-input', parms.get('lengthwidthratio-max-input', 0))

# to scroll down
for i in range(50):
    mult_tab = br.find_elements_by_css_selector('a[class="grid-row row TL511DiaStrikePrice"]')
    br.execute_script("arguments[0].scrollIntoView();", mult_tab[-1])
    if wait_tag(br, 2, By.CLASS_NAME, 'grid-alert.show-alert'):
        print('to the bottom of the the page')
        break

# to get the contents list
body = BeautifulSoup(br.page_source, 'lxml')
alist = body.find_all(name='a', attrs={'class': 'grid-row row TL511DiaStrikePrice'})  # get all 
new_list = []
for a in alist:
    shape = fs(a.find(name='div', attrs={'class': 'row-cell shape'}).text)
    price = fs(a.find(name='div', attrs={'class': 'row-cell price'}).text)
    carat = fs(a.find(name='div', attrs={'class': 'row-cell carat'}).text)
    cut = fs(a.find(name='div', attrs={'class': 'row-cell cut'}).text)
    color = fs(a.find(name='div', attrs={'class': 'row-cell color'}).text)
    clarity = fs(a.find(name='div', attrs={'class': 'row-cell clarity'}).text)
    polish = fs(a.find(name='div', attrs={'class': 'row-cell polish'}).text)
    symmetry = fs(a.find(name='div', attrs={'class': 'row-cell symmetry'}).text)
    fluorescence = fs(a.find(name='div', attrs={'class': 'row-cell fluorescence'}).text)
    depth = fs(a.find(name='div', attrs={'class': 'row-cell depth'}).text)
    table = fs(a.find(name='div', attrs={'class': 'row-cell table'}).text)
    lxwratio = fs(a.find(name='div', attrs={'class': 'row-cell lxwratio'}).text)
    pricepercarat = fs(a.find(name='div', attrs={'class': 'row-cell pricepercarat'}).text)
    culet = fs(a.find(name='div', attrs={'class': 'row-cell culet'}).text)
    date = fs(a.find(name='div', attrs={'class': 'row-cell date'}).text)
    # print('{}-{}-{}-{}-{}-{}-{}-{}-{}-{}-{}-{}-{}-{}-{}'.format(shape, price, carat, cut, color, clarity, polish,
                                                        #        symmetry, fluorescence, depth, table, lxwratio,
                                                      #          pricepercarat, culet, date))
    new_list.append([shape, price, carat, cut, color, clarity, polish,
                     symmetry, fluorescence, depth, table, lxwratio,
                     pricepercarat, culet, date])

# creat data frame
df = pd.DataFrame(new_list)
df.rename(columns={0: 'shape', 1: 'price', 2: 'carat', 3: 'cut', 4: 'color', 5: 'clarity', 6: 'polish',
                   7: 'symmetry', 8: 'fluorescence', 9: 'depth', 10: 'table', 11: 'lxwratio',
                   12: 'pricepercarat',
                   13: 'culet', 14: 'date'}, inplace=True)




if __name__ == "__main__":
    pass
```
